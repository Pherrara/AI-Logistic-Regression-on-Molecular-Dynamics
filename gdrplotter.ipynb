```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.style.use(['science','notebook','grid'])
```

```python
x,y = [], []

with open("/home/pher/Documents/computational/dinamica_molecolare_work/2023/ro0.3_t1/gdrfile",'r') as f:
    for l in f:
        row = l.split()
        x.append(float(row[0]))
        y.append(float(row[1]))
```

```python
plt.figure(figsize=(16,9)) #must come BEFORE the actual plot
plt.scatter(x, y, color="red", label="data")
#plt.plot(x_model,y_model, color='r', label="fit")
#plt.legend(loc='best', fontsize=12, ncol=2) #ncol is the number of column of the legend

#props = dict(boxstyle='square', facecolor='white', alpha=0.9)
#my_string = r'$a \cdot \cos ^{2}(\theta)+c$' + "\n" + r'$a$' + " = " + str(a_opt) + r' $\pm$ ' + str(a_err) + "\n" + r'$c$' + " = " + str(c_opt) + r' $\pm$ ' + str(c_err)
#plt.text(-20, 0.3, my_string, fontsize=15, bbox=props)

#plt.locator_params(y=20) this is about the density of thicks on both axis I could add <, axis='x'> inside the parentesis, to do just for one axis
plt.xlabel('radius (sigma units)')
plt.ylabel('g(r)')
plt.title('GDR')
plt.yscale('linear')
plt.xscale('linear')
#plt.savefig('manual_histogram_plot.jpg') #always save before plt.show() it should be "namefile.png" or "namefile.pdf"
plt.show()
```
